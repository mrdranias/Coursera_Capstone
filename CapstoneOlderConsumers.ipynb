### Aim and Target Audience
This report aims to use machine learning and data available from Foursquare to predict which shops and neighborhoods might appeal to older consumers. The resulting model is of interest to people producing products that target this market. The model predictions will enable these producers to identify locations for shops, advertising, and product placement

# Data 
### Data Sources and Cleaning
Demographic data for 7 cities in the state of Florida were pulled from wikipedia, http://zipatlas.com/us/fl/city-comparison/median-age.5.htm and the US Census https://factfinder.census.gov/faces/nav/jsf/pages/index.xhtml. When data disagrees, US census data is used. Information regarding venues in each of these cities were pulled using the Foursquare API. 

### Selection of City Sample
Tables of median ages in Florida cities from the websites listed above were examined. The aim was to find 3 "older" cities and 3 "younger" cities. Cities had to have more than 20000 residents to be considered. A table of 7 cities was ultimately constructed: 3 training cities with a median age higher than 45 (Boca Raton, Delray Beach, Ormond Beach mean age=51); 3 training cities with a median age lower than 35 (Gainesville, Tallahassee, Homestead mean age = 30) ; 1 test city of intermediate age and high familiarity (Orlando, median age 33). Metro Orlando consists of 15 zip codes and was selected to be the test case as the author lives locally in Orlando and has familiarity to help establish the ground truth of shopper demographics for these different zip codes. Specifically for each zip code 2 truth values are selected, one from census data for age and the other from personal observation as a native of Orlando.

### Computation of Venue Feature Vectors

Venue category frequencies are computed for each city and zip code and used to construct a common venue feature vector. In order to construct a common feature vector, feature extraction had to be done. The easiest thing was to build higher order venue categories such as Restaurants, Bars, Healthcare, Shopping, Foodstores, Hobbies, Apparel, Entertainment. To avoid sparse data with a small dataset, venue categories mentioned appearing only in one city or zip code data set were dropped. Venue feature vectors are normalized by the total number of venues reported for each city or zip code. Final feature vectors have normalized frequencies across 62 venue categories. For Specific code see Appendix: [4. Compute Venue Feature Vector](#4.-Compute-Venue-Feature-Vector)

# Methodology 

### Exploratory data Analysis
Cities are plotted on a map and cross correlation matrices are computed to visualize the similarity between different city venue feature vectors. Unsupervised clustering
(a hierarchical clustering/linkage algorithm) is used to order the cities and zip codes so that the natural clusters of similar cities are visualized. The relative frequencies of different categories of venue were also plotted across younger and older cities. The features found to vary most were 'Apparel','Healthcare','Hotels','Bank','Sports','Antique Shop']]. Limiting the cross-correlation and clustering to these dimensions led to the formation of 2 unsupervised clusters that roughly corresponded to young (Gainesville, Tallahassee, Orlando, Delray Beach) and old (Boca Raton, Ormond Beach, Homestead) except that Homestead and Delray Beach were 'incorrectly' clustered for age. As a consequence a nonlinear classifier is used to segregate old and young cities. For Specific code see Appendix: [5. Data Exploration](#5.-Data-Exploration),

### Age Classification of Zipcodes.
Two classifiers were probed to find a best and simplest model for classifying young vs. old cities: KNN and SVM. The training data was the 6 cities with dimensions limited to the 6 described above. The testing data was zipcodes in Orlando with correct answers for young vs. old set to a threshold of 35 and younger or 36 and older. This results in a chance classification rate of 50%. Gridsearch was used to explore classifiers hyperparameters. Data was standard scaled prior to classification. For Specific code see Appendix:[6. Machine Learning Classification of Cities and Zipcodes](#6.-Machine-Learning-Classification-of-Cities-and-Zipcodes)

# Results 
### Classifier Results
The best KNN classifier paid attention to 2 nearest neighbors, had a distance metric and a training classification of 100%. KNN classifier testing classification on zipcodes using demographic median age of 35 as a cutoff resulted in 60% accuracy where chance was 50%. The best SVM classifier was an RBF kernel with a training classification was 100% and a testing classification of 70%. The anomalous zipcodes for the SVM are: 32789,32804,32807,32822,32824,32832. There was no bias in mistakes: half are from each age class.  The location of these zipcodes was plotted on a map of Orlando but no obvious geographical reason for the errors was seen. Review of specific misclassifications suggests the presence of healthcare and sports facilities in younger zip codes results in misclassification as old (32822,32832). The presence of hotels seems to be primarily responsible for older neighborhoods  being classified as young (32804,32824). Of the remaining errors We suggest that 32789 is actually accurately classified as Older in terms of its consumer profile even though its demographics trend younger. For Specific Code and figures see Appendix: [7. Analysis of Results](#-7.-Analysis-of-Results)

# Discussion 
The rationale of looking at younger and older cities for shopping patterns was to control for mobility within a city. Within a city demographics at the zip code level may reveal very little about where locals are doing their shopping. However even when zip code demographics are taken as the gold standard for the model, model results were surprisingly accurate: an accuracy of 70% on 20 zipcodes. Of the errors, the one that was most supporting of the premise was the identification of zip code 32789 as an "older" area. By demographic standards it was misclassified but by shopping standards it definitely draws an older crowd. 32789 is Winter Park and its basically where the wealthier retirees go to party and shop. The demographics are probably an artifact of Rollins College which is situated in the zip code but the students likely participate very little in the local economy. 
Some of the more general findings were intuitive: older neighborhoods have banks. Younger neighborhoods may not. Older neighborhoods have more antique shops. Some obvious improvements would be to expand the number of cities involved so that classifiers with statistical power could be meaningfully used. The data cries out for a multiple logistic regression analysis. 


# Conclusion 
The goal here is to identify neighborhoods where older people live and perhaps prefer to shop. Some affirmative results were found that would justify developing the model. The aim was to use cities with older ages

###  2. Use Foursquare to pull data on venues in 7 cities.

```python
# The code was removed by Watson Studio for sharing.
```

```python
#2. Pull foursquare venue information on the cities
city_list=['Gainesville', 'Tallahassee', 'Homestead', 'Boca Raton', 'Delray Beach', 'Ormond Beach', 'Orlando' ]

# * define query url stem
citywide_url_query_stem = 'https://api.foursquare.com/v2/venues/search?near={},{}&client_id={}&client_secret={}&v={}&limit={}'

#* specify query parameter tuple (CITY, STATE, CLIENT_ID, CLIENT_SECRET, VERSION, LIMIT)
LIMIT =250 
VERSION = '20180604'
STATE= 'FL'

#Make dictonary to store venue data for each city
records_dict = dict.fromkeys(city_list)

#loop through cities and extract venue data
for idx, CITY in enumerate(city_list):
    url_query = citywide_url_query_stem.format(CITY,STATE, CLIENT_ID, CLIENT_SECRET, VERSION, LIMIT)
    records_dict[CITY] = requests.get(url_query).json()
```

```python
# 3. extract venue name, venue category information, venue category frequency for each city 
#* call get_category_type function
# function that extracts the category of the venue
def get_category_type(row):
    try:
        categories_list = row['categories']
    except:
        categories_list = row['venue.categories']
        
    if len(categories_list) == 0:
        return None
    else:
        return categories_list[0]['name']

# * iterate through JSON dictionary, assign to dataframe dictionary.
dataframe_dict = dict.fromkeys(city_list)

for CITY in city_list:
    #readout category relevant foursquare data.
    items = records_dict[CITY]['response']['venues']
    
    #Run NY JSON data extraction routine
    # flatten JSON
    dataframe = json_normalize(items)
    # filter columns
    filtered_columns = ['name', 'categories']
    dataframe_filtered = dataframe.loc[:, filtered_columns]
    # filter the category for each row
    dataframe_filtered['categories'] = dataframe_filtered.apply(get_category_type, axis=1)
    # clean columns
    dataframe_filtered.columns = [col.split('.')[-1] for col in dataframe_filtered.columns]
    dataframe_filtered['city']=CITY
    #extract (city, name, categories) assign to dataframe dictionary.
    dataframe_dict[CITY] = dataframe_filtered[['city','name','categories']]
```

```python
#3. Clean dictionary of dataframes and Display top 3 entries for each
for idx,CITY in enumerate(city_list):
    #standardize fields so easy to merge/join
    dataframe_dict[CITY] = dataframe_dict[CITY].rename({'city':'City','name':'venue', 'categories':'Category'},axis=1)   
    #make index orderly
    dataframe_dict[CITY] = dataframe_dict[CITY].reset_index(drop=True)
    #Put postcode first
    dataframe_dict[CITY]= dataframe_dict[CITY].loc[:,['City','venue','Category']]
    #c. display top 3 entries for each postal code
    print("{}. First 3 Venues for city {}".format(idx+1, CITY))
    display(HTML(dataframe_dict[CITY].head(3).to_html()))
    print()
```

### 3. Extract latitude and longitude for Orlando zipcodes and query Foursquare
    1. Copy website for zipcodes https://gis.orlando.gov/PDF_Docs/CityWideMaps/ZipCodeMap8x11.pdf
    2. Get longitude and latitude for each zip code 
    3. Pull foursquare venue information on the postal codes.


```python
#1. Get Orlando Zip Codes
zipcodes=[32789,32801,32803,32804,32805,32806,32807,32808,32809,32810,32811,32812,32819,32822,32824,32827,32829,32832,32835,32839]
OrlandoZip_df = pd.DataFrame(index=zipcodes).reset_index()
OrlandoZip_df = OrlandoZip_df.rename({'index':'ZipCode'},axis=1)   

#2 Get longitude and latitude for each zip code
geolocator = Nominatim(user_agent="toronto_explorer")
for idx,ZIP in enumerate(zipcodes):
    location = geolocator.geocode(ZIP)
    latitude = location.latitude
    longitude = location.longitude
    OrlandoZip_df.loc[idx,'latitude']=latitude
    OrlandoZip_df.loc[idx,'longitude']=longitude

OrlandoZip_df.T
```

```python
#3. Pull foursquare venue information on the postal codes.
#b Query foursquare for venues by location (lat and long of top5)
# * define query url stem
url_query_stem = 'https://api.foursquare.com/v2/venues/explore?client_id={}&client_secret={}&ll={},{}&v={}&radius={}&limit={}'

#* specify query parameter tuple (CLIENT_ID, CLIENT_SECRET, latitude, longitude, VERSION, radius, LIMIT)
LIMIT =200 
radius = 5000
VERSION = '20180604'

# * call for top5, record results in dictionary.
Orlando_records_dict = dict.fromkeys(OrlandoZip_df['ZipCode'])
for idx, ZIP in enumerate(OrlandoZip_df['ZipCode']):
    latitude = OrlandoZip_df.loc[idx,'latitude']
    longitude = OrlandoZip_df.loc[idx,'longitude']
    url_query = url_query_stem.format(CLIENT_ID, CLIENT_SECRET, latitude, longitude, VERSION, radius, LIMIT)
    Orlando_records_dict[ZIP] = requests.get(url_query).json()
```

```python
# * iterate through JSON dictionary, assign to dataframe dictionary.
Zip_dataframe_dict = dict.fromkeys(zipcodes)

for idx,ZIP in enumerate(zipcodes):
    #readout category relevant foursquare data.
    items =Orlando_records_dict[ZIP]['response']['groups'][0]['items']
    
    #Run NY JSON data extraction routine
    # flatten JSON
    dataframe = json_normalize(items)
    # filter columns
    filtered_columns = ['venue.name', 'venue.categories'] + [col for col in dataframe.columns if col.startswith('venue.location.')] + ['venue.id']
    dataframe_filtered = dataframe.loc[:, filtered_columns]
    # filter the category for each row
    dataframe_filtered['venue.categories'] = dataframe_filtered.apply(get_category_type, axis=1)
    # clean columns
    dataframe_filtered.columns = [col.split('.')[-1] for col in dataframe_filtered.columns]

    #extract (city, name, categories) assign to dataframe dictionary.
    Zip_dataframe_dict[ZIP] = dataframe_filtered[['name','postalCode','categories','lat','lng']]
```

```python
#3. Clean dictionary of dataframes and Display top 3 entries for each
for idx,ZIP in enumerate(zipcodes):
    #standardize fields so easy to merge/join
    Zip_dataframe_dict[ZIP] = Zip_dataframe_dict[ZIP].rename({'postalCode':'ZipCode','name':'venue', 'categories':'Category'},axis=1)   
    #make index orderly
    Zip_dataframe_dict[ZIP] = Zip_dataframe_dict[ZIP].reset_index(drop=True)
    #Put postcode first
    Zip_dataframe_dict[ZIP]= Zip_dataframe_dict[ZIP].loc[:,['ZipCode','venue','Category']]
    #Clean zipcode: drop nans, drop zip suffix
    Zip_dataframe_dict[ZIP].dropna(how='any', axis=0,inplace=True)    
    Zip_dataframe_dict[ZIP].loc[:,'ZipCode'] = Zip_dataframe_dict[ZIP].loc[:,'ZipCode'].apply(lambda x: str(x)[:5])
    Zip_dataframe_dict[ZIP].reset_index(drop=True,inplace=True)
    
    #c. display top 3 entries for each postal code
    print("{}. First 3 Venues for zipcode {}".format(idx+1, ZIP))
    display(HTML(Zip_dataframe_dict[ZIP].head(3).to_html()))
    print()
```

### 4. Compute Venue Feature Vector
    1. one hot encode city and zip code venue information 
    2. Extract higher order categories 
    3. Compute normalized venue vectors.


```python
#1. one hot encode city and zip code venue information
#compile city dataframe
city_df = dataframe_dict[city_list[0]]
for CITY in city_list[1:]:
    city_df = city_df.append(dataframe_dict[CITY], ignore_index=True)
    
#compile zipcode data frame
zip_df = Zip_dataframe_dict[zipcodes[0]]
for ZIP in zipcodes[1:]:
    zip_df = zip_df.append(Zip_dataframe_dict[ZIP], ignore_index=True)

# use one hot encoding on city data and zip code data to create binary list of venu categories
raw_city_vector = pd.get_dummies(city_df.set_index('City')[['Category']], prefix="", prefix_sep="")
raw_zip_vector = pd.get_dummies(zip_df.set_index('ZipCode')[['Category']], prefix="", prefix_sep="")

#clean out extra zipcodes
raw_zip_vector = raw_zip_vector.loc[[str(x) for x in zipcodes],:]
```

```python
#2. Extract higher order categories 
zip_vector = raw_zip_vector.groupby('ZipCode').sum()
city_vector = raw_city_vector.groupby('City').sum()

#Group restaurants
lump_restaurants = [x for x in city_vector.columns if x.endswith('Restaurant')]
lump_restaurants.extend([x for x in zip_vector.columns if x.endswith('Restaurant')])
lump_restaurants.extend(['Fish & Chips Shop','Hot Dog Joint', 'Food & Drink Shop','Noodle House','Salad Place','Taco Place','Steakhouse'])
lump_restaurants.extend(['Wings Joint', 'BBQ Joint','Food & Drink Shop','Deli / Bodega', 'Coffee Shop','Food Truck','Bistro','Café', 'Diner',
                         'Sandwich Place','Tea Room','Pizza Place','Burger Joint','Breakfast Spot','Street Food Gathering','Bagel Shop',
                        'Fried Chicken Joint', 'Burrito Place'])
drop_city = list(set(city_vector.columns.tolist()).intersection(set(lump_restaurants)))
drop_zip = list(set(zip_vector.columns.tolist()).intersection(set(lump_restaurants)))

city_vector['Restaurants']=city_vector[drop_city].sum(axis=1)
city_vector.drop(columns = drop_city, inplace=True)
zip_vector['Restaurants']=zip_vector[drop_zip].sum(axis=1)
zip_vector.drop(columns = drop_zip, inplace=True)

# Group museums and galleries
lump_museums = [x for x in city_vector.columns if x.endswith('Museum')|x.endswith('Gallery')|x.endswith('Library')]
lump_museums.extend([x for x in zip_vector.columns if x.endswith('Museum')|x.endswith('Gallery')|x.endswith('Library')])
lump_museums.extend(['Aquarium','Exhibit','Sculpture Garden','Zoo'])
lump_museums=list(set(lump_museums))

drop_city = list(set(city_vector.columns.tolist()).intersection(set(lump_museums)))
drop_zip = list(set(zip_vector.columns.tolist()).intersection(set(lump_museums)))

city_vector['Museums']=city_vector[drop_city].sum(axis=1)
city_vector.drop(columns = drop_city, inplace=True)
zip_vector['Museums']=zip_vector[drop_zip].sum(axis=1)
zip_vector.drop(columns = drop_zip, inplace=True)

# Group colleges
lump_colleges = [x for x in city_vector.columns if x.endswith('College')|x.startswith('College')]
lump_colleges.extend([x for x in zip_vector.columns if x.endswith('College')|x.startswith('College')])
lump_colleges.extend(['General College & University','Medical School'])
lump_colleges=list(set(lump_colleges))

drop_city = list(set(city_vector.columns.tolist()).intersection(set(lump_colleges)))
drop_zip = list(set(zip_vector.columns.tolist()).intersection(set(lump_colleges)))

city_vector['Colleges']=city_vector[drop_city].sum(axis=1)
city_vector.drop(columns = drop_city, inplace=True)
zip_vector['Colleges']=zip_vector[drop_zip].sum(axis=1)
zip_vector.drop(columns = drop_zip, inplace=True)
                     
# Group Bars
lump_bars = [x for x in city_vector.columns if x.endswith('Bar')|x.startswith('Bar')]
lump_bars = [x for x in zip_vector.columns if x.endswith('Bar')|x.startswith('Bar')]
lump_bars.extend( [x for x in city_vector.columns if x.endswith('Pub')|x.startswith('Pub')])
lump_bars.extend(['Brewery','Speakeasy', 'Winery','Wine Shop', 'Lounge', 'Beer Store','Liquor Store','Beer Garden','Gastropub'])
lump_bars=list(set(lump_bars))

drop_city = list(set(city_vector.columns.tolist()).intersection(set(lump_bars)))
drop_zip = list(set(zip_vector.columns.tolist()).intersection(set(lump_bars)))

city_vector['Bars']=city_vector[drop_city].sum(axis=1)
city_vector.drop(columns = drop_city, inplace=True)
zip_vector['Bars']=zip_vector[drop_zip].sum(axis=1)
zip_vector.drop(columns = drop_zip, inplace=True)
#2. Extract higher order categories continued            
#group food stores
lump_food=['Supplement Shop', 'Warehouse Store', 'Food', 'Farmers Market', 'Fish Market','Grocery Store','Gourmet Shop', 
           'Fruit & Vegetable Store','Butcher'] 

drop_city = list(set(city_vector.columns.tolist()).intersection(set(lump_food)))
drop_zip = list(set(zip_vector.columns.tolist()).intersection(set(lump_food)))
city_vector['Foodstore']=city_vector[drop_city].sum(axis=1)
city_vector.drop(columns = drop_city, inplace=True)
zip_vector['Foodstore']=zip_vector[drop_zip].sum(axis=1)
zip_vector.drop(columns = drop_zip, inplace=True)
                      
#group apparel stores
lump_clothes=["Women's Store",'Shoe Store', 'Jewelry Store','Accessories Store','Clothing Store','Boutique',
              'Lingerie Store','Cosmetics Shop','Thrift / Vintage Store','Baby Store','Kids Store']
                     
drop_city = list(set(city_vector.columns.tolist()).intersection(set(lump_clothes)))
drop_zip = list(set(zip_vector.columns.tolist()).intersection(set(lump_clothes)))
city_vector['Apparel']=city_vector[drop_city].sum(axis=1)
city_vector.drop(columns = drop_city, inplace=True)
zip_vector['Apparel']=zip_vector[drop_zip].sum(axis=1)
zip_vector.drop(columns = drop_zip, inplace=True)

#books and hobbies                
lump_hobbies = ['Newsstand','Sporting Goods Shop','Hobby Shop','Bike Shop','Arts & Crafts Store','Comic Shop',
                'Paper / Office Supplies Store','Bookstore','Video Store','Video Game Store']                  

drop_city = list(set(city_vector.columns.tolist()).intersection(set(lump_hobbies)))
drop_zip = list(set(zip_vector.columns.tolist()).intersection(set(lump_hobbies)))
city_vector['Hobby&BookShop']=city_vector[drop_city].sum(axis=1)
city_vector.drop(columns = drop_city, inplace=True)
zip_vector['Hobby&BookShop']=zip_vector[drop_zip].sum(axis=1)
zip_vector.drop(columns = drop_zip, inplace=True)
                      
#group other shops
lump_shops = ['Record Shop', 'Shopping Mall', 'Discount Store','Pet Store', 'Music Store', 'Fabric Shop','Gift Shop',
              'Flower Shop','Big Box Store','Electronics Store', 'Furniture / Home Store','Toy / Game Store',
              'Adult Boutique','Miscellaneous Shop','Outlet Mall','Department Store','Hardware Store']                 

drop_city = list(set(city_vector.columns.tolist()).intersection(set(lump_shops)))
drop_zip = list(set(zip_vector.columns.tolist()).intersection(set(lump_shops)))
city_vector['MiscShop']=city_vector[drop_city].sum(axis=1)
city_vector.drop(columns = drop_city, inplace=True)
zip_vector['MiscShop']=zip_vector[drop_zip].sum(axis=1)
zip_vector.drop(columns = drop_zip, inplace=True)
                      
#desserts
lump_sweets= ['Frozen Yogurt Shop','Candy Store', 'Chocolate Shop','Donut Shop','Ice Cream Shop', 'Bakery','Dessert Shop','Cupcake Shop',
              'Smoothie Shop','Bubble Tea Shop']
                     
drop_city = list(set(city_vector.columns.tolist()).intersection(set(lump_sweets)))
drop_zip = list(set(zip_vector.columns.tolist()).intersection(set(lump_sweets)))
city_vector['Sweetshop']=city_vector[drop_city].sum(axis=1)
city_vector.drop(columns = drop_city, inplace=True)
zip_vector['Sweetshop']=zip_vector[drop_zip].sum(axis=1)
zip_vector.drop(columns = drop_zip, inplace=True)
              
# group healthcare
lump_health=['Veterinarian','Chiropractor','Hospital','Acupuncturist',"Doctor's Office","Dentist's Office",'Pharmacy', 'Medical Center']

drop_city = list(set(city_vector.columns.tolist()).intersection(set(lump_health)))
drop_zip = list(set(zip_vector.columns.tolist()).intersection(set(lump_health)))
city_vector['Healthcare']=city_vector[drop_city].sum(axis=1)
city_vector.drop(columns = drop_city, inplace=True)
zip_vector['Healthcare']=zip_vector[drop_zip].sum(axis=1)
zip_vector.drop(columns = drop_zip, inplace=True)

# group sports 
lump_sports=['Spa','Soccer Field','Gymnastics Gym','Gym / Fitness Center', 'Mini Golf', 'Yoga Studio', 'Gym','Skating Rink',
             'Basketball Court','Golf Course','Pool','Paintball Field']

drop_city = list(set(city_vector.columns.tolist()).intersection(set(lump_sports)))
drop_zip = list(set(zip_vector.columns.tolist()).intersection(set(lump_sports)))
city_vector['Sports']=city_vector[drop_city].sum(axis=1)
city_vector.drop(columns = drop_city, inplace=True)
zip_vector['Sports']=zip_vector[drop_zip].sum(axis=1)
zip_vector.drop(columns = drop_zip, inplace=True)


# group non-college schools
lump_schools = [x for x in city_vector.columns if x.endswith('School')|x.startswith('School')]
lump_schools.extend([x for x in zip_vector.columns if x.endswith('School')|x.startswith('School')])

drop_city = list(set(city_vector.columns.tolist()).intersection(set(lump_schools)))
drop_zip = list(set(zip_vector.columns.tolist()).intersection(set(lump_schools)))
city_vector['Schools']=city_vector[drop_city].sum(axis=1)
city_vector.drop(columns = drop_city, inplace=True)
zip_vector['Schools']=zip_vector[drop_zip].sum(axis=1)
zip_vector.drop(columns = drop_zip, inplace=True)


#lump entertainment venues
lump_entertainment=['Basketball Stadium','Comedy Club','Movie Theater','Performing Arts Venue','Soccer Stadium',
                    'Theme Park Ride / Attraction','Theme Park','Concert Hall','Event Space','General Entertainment','Nightclub',
                   'Rock Club','Concert Hall','Social Club']

drop_city = list(set(city_vector.columns.tolist()).intersection(set(lump_entertainment)))
drop_zip = list(set(zip_vector.columns.tolist()).intersection(set(lump_entertainment)))
city_vector['Entertainment']=city_vector[drop_city].sum(axis=1)
city_vector.drop(columns = drop_city, inplace=True)
zip_vector['Entertainment']=zip_vector[drop_zip].sum(axis=1)
zip_vector.drop(columns = drop_zip, inplace=True)

#lump auto and travel
lump_travel=['Auto Dealership', 'Automotive Shop','Bus Line', 'Bus Station','General Travel', 'Motorcycle Shop','Rental Car Location','Airport']

drop_city = list(set(city_vector.columns.tolist()).intersection(set(lump_travel)))
drop_zip = list(set(zip_vector.columns.tolist()).intersection(set(lump_travel)))
city_vector['Travels']=city_vector[drop_city].sum(axis=1)
city_vector.drop(columns = drop_city, inplace=True)
zip_vector['Travels']=zip_vector[drop_zip].sum(axis=1)
zip_vector.drop(columns = drop_zip, inplace=True)

#lump hotels
lump_hotel = [ 'Bed & Breakfast','Resort','Campground','Hotel']

drop_city = list(set(city_vector.columns.tolist()).intersection(set(lump_hotel)))
drop_zip = list(set(zip_vector.columns.tolist()).intersection(set(lump_hotel)))
city_vector['Hotels']=city_vector[drop_city].sum(axis=1)
city_vector.drop(columns = drop_city, inplace=True)
zip_vector['Hotels']=zip_vector[drop_zip].sum(axis=1)
zip_vector.drop(columns = drop_zip, inplace=True)

#drop all city columns that contain only 1 response
columns_to_drop = [x for x in city_vector.columns if city_vector.loc[:,x].sum()<2]
city_vector.drop(columns =city_vector[columns_to_drop].columns.tolist(), inplace=True)

#drop all zip columns that contain only 2 response
columns_to_drop = [x for x in zip_vector.columns if zip_vector.loc[:,x].sum()<2]
zip_vector.drop(columns =zip_vector[columns_to_drop].columns.tolist(), inplace=True)

#drop items not in city vector
columns_to_drop = list(set(zip_vector.columns.tolist()).difference(city_vector.columns.tolist()))
zip_vector.drop(columns =zip_vector[columns_to_drop].columns.tolist(), inplace=True)

#add columns in the city vector that are not in the zip vector (and set to zero)
city_columns = list(set(city_vector.columns.tolist()).difference(zip_vector.columns.tolist()))
empty_city = pd.DataFrame(index=[str(x) for x in zipcodes], columns=city_columns)
zip_vector2 = pd.concat([zip_vector,empty_city],axis=0,sort=True)
zip_vector = zip_vector2.iloc[:len(zip_vector),:]
zip_vector.fillna(0, inplace=True)
zip_vector
```

### 6. Machine Learning Classification of Cities and Zipcodes
    1. format citywide feature and target vectors
    2. format zipcode test data
    3. KNN Classifier best results 
    4. SVM Classifier best results


```python
#1. format city and zipcode features limited to strongly shared, covarying features
y_prime_city = pd.get_dummies(norm_city_vector.dropna(how='any',axis=0).loc[:,'Age']).iloc[:,0]
norm_city_X_df = norm_city_vector.dropna(how='any',axis=0).loc[:,['Apparel','Healthcare','Hotels','Bank','Sports','Antique Shop']]
X_city = preprocessing.StandardScaler().fit(norm_city_X_df).transform(norm_city_X_df)
y_prime_city
```

```python
#2. Format test data: zipcodes
#classify as older (36+)/younger (-35
#[Old, Old, old, Old, old, old, Young, Young, Young, Old,Old, Young, old,old,Young,Young,Young,Young])
# [44,36,39,41,37,38,35,32,35,33,31,39,40,35,36,38,34,33,33,31]
norm_zip_vector.loc[:,'Age']=pd.Series([44,36,39,41,37,38,35,32,35,33,31,39,40,35,36,38,34,33,33,31])
norm_zip_vector.loc[:,'Age']=pd.DataFrame( [44,36,39,41,37,38,35,32,35,33,31,39,40,35,36,38,34,33,33,31], index=norm_zip_vector.index,columns=['Age'])
norm_zip_vector.loc[:,'AgeRank']=norm_zip_vector.loc[:,'Age'].apply(lambda x: np.sign(np.sign(x-36)+1))
norm_zip_vector.loc[:,'AgeRank2']=norm_zip_vector.loc[:,'Age'].apply(lambda x: np.sign(np.sign(x-35)+1))

y_test_36 = pd.get_dummies(norm_zip_vector.dropna(how='any',axis=0).loc[:,'AgeRank']).iloc[:,0]
y_test_35 = pd.get_dummies(norm_zip_vector.dropna(how='any',axis=0).loc[:,'AgeRank2']).iloc[:,0]

norm_zip_X_df = norm_zip_vector.dropna(how='any',axis=0).loc[:,['Apparel','Healthcare','Hotels','Bank','Sports','Antique Shop']]
X_test = preprocessing.StandardScaler().fit(norm_zip_X_df).transform(norm_zip_X_df)
```

```python
#3. Train KNN classifier and Grid Search for parameters
#declare classifier
KNC = KNeighborsClassifier()

#declare parameters for grid search
grid_param= {'n_neighbors': [2, 3],
             'weights': ["uniform", "distance"], 
             'p': [1,2]}


#declare grid search parameters
grid_knc = GridSearchCV(estimator=KNC,
                        param_grid=grid_param,
                        scoring= 'f1',
                        n_jobs=-1,
                        return_train_score=True,
                       )

#execute grid search
grid_knc.fit(X_city, y_prime_city)

#report best parametric case
best_parameters = grid_knc.best_params_  
print(best_parameters)
print("\n" + " the validation f1 score is: %.4f" %grid_knc.best_score_)

#report scores for training data overall.
y_hat = grid_knc.predict(X_city)
KNN_trainf1 = f1_score(y_prime_city,y_hat)


# New Model Evaluation metrics 
print('Overall Training Accuracy Score : ' + str(accuracy_score(y_prime_city,y_hat)))
print('Overall Training Precision Score : ' + str(precision_score(y_prime_city,y_hat)))
print('Overall Training Recall Score : ' + str(recall_score(y_prime_city,y_hat)))
print('Overall Training F1 Score : ' + str(f1_score(y_prime_city,y_hat)))
```

```python
#print out test data scoring with old=36 and above
#Predict test values based on new parameters
print("\n" + "if old rank is set to 36 (chance=50%):")
y_hat_test = grid_knc.predict(X_test)

KNN_test_f1 = f1_score(y_test_36,y_hat_test)
# New Model Evaluation metrics 
print('Testing Accuracy Score : ' + str(accuracy_score(y_test_36,y_hat_test)))
print('Testing Precision Score : ' + str(precision_score(y_test_36,y_hat_test)))
print('Testing Recall Score : ' + str(recall_score(y_test_36,y_hat_test)))
print('Testing F1 Score : ' + str(f1_score(y_test_36,y_hat_test)))


print("\n" + "Validation f1 score is %.4f, test f1 score is %.4f" %(KNN_trainf1,
                                                                    KNN_test_f1))

```

```python
#4. Train SVM classifier and Grid Search for parameters
# declare classifier, define parameters, declare gridSearch, run gridSearch, print best.

# declare classifier instance
SVM2 = SVC()

#declare parameters for search (dictionary of lists)
grid_param={'C' : [0.5, 1.0, 2.0, 5.0, 10.0, 100.0], 
            'kernel': ['linear','rbf', 'poly', 'sigmoid'],
            'degree': [2, 3, 4],
            'shrinking' : [True, False],
           }

# declare grid search object
grid_svm = GridSearchCV(estimator = SVM2,
                       param_grid = grid_param,
                       scoring = 'f1',
                       verbose=True)

#execute grid search
grid_svm.fit(X_city, y_prime_city)

#report best parametric case
best_parameters = grid_svm.best_params_  
print(best_parameters)
#Print out training data scoring and perhaps validation set.
#report best accuracy?
SVM_trainf1 = grid_svm.best_score_
print("The validation set f1 is {0:2f}".format(SVM_trainf1))

#report scores for training data overall.
y_hat3 = grid_svm.predict(X_city)


# New Model Evaluation metrics 
print('Overall Training Accuracy Score : ' + str(accuracy_score(y_prime_city,y_hat3)))
print('Overall Training Precision Score : ' + str(precision_score(y_prime_city,y_hat3)))
print('Overall Training Recall Score : ' + str(recall_score(y_prime_city,y_hat3)))
print('Overall Training F1 Score : ' + str(f1_score(y_prime_city,y_hat3)))
```
